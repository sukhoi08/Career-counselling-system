```python
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('/content/career_guidance.csv')
```

```python
from sklearn.preprocessing import OrdinalEncoder
entrep_aspirations_order = [['Low', 'Medium', 'High']]
encoder = OrdinalEncoder(categories=entrep_aspirations_order)
encoded_data = encoder.fit_transform(df[['Entrepreneurial_Aspirations']])
df['Entrep_encoded'] = encoded_data
employment_order = [['Internship', 'Part-time', 'Full-time']]
encoder2 = OrdinalEncoder(categories=employment_order)
encoded_data2 = encoder2.fit_transform(df[['Employment_Type']])
df['Emp_encoded'] = encoded_data2
df.drop(columns=['Employment_Type', 'Entrepreneurial_Aspirations'], inplace=True)
```

```python

print("pandas", pd.__version__)
print("scikit-learn", sklearn.__version__)
```

```python
y = df['Recommended_Career_Path']  # Target variable
X = df.drop(columns=['Recommended_Career_Path'])
```

```python
X.columns
```

```python
columns_to_encode = ['Gender', 'Field_of_Study', 'Career_Interests', 'University_Location']
X_encoded = pd.get_dummies(X, columns=columns_to_encode, prefix=columns_to_encode)
```

```python
X = X_encoded
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X['Age'] = scaler.fit_transform(X[['Age']])
X['GPA'] = scaler.fit_transform(X[['GPA']])
```

```python
y_encoded_data = pd.get_dummies(y)
```

```python
y = y_encoded_data
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import joblib
joblib.dump(list(X_train.columns), 'feature_names.pkl')
```

```python
import pickle
with open('career_path_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
